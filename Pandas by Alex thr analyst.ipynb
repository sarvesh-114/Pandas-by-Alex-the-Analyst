## **Merging, joining and Concatenating dataFrames**

```python
data1 = pd.read_csv(r"C:\Users\sarvesh jathar\Downloads\LOTR.csv")
data1.head()
```

```python
data2 = pd.read_csv(r"C:\Users\sarvesh jathar\Downloads\LOTR 2.csv")
data2.head()
```

Like joins in SQL

```python
data1.merge(data2, how = "inner")
```

```python
data1.merge(data2, how = "inner", on = ["FellowshipID", "FirstName"])
```

```python
data1.merge(data2, how = "left")
```

```python
data1.merge(data2, how = "right")
```

```python
data1.merge(data2, how = "outer")
```

```python
 data1.set_index("FellowshipID", inplace = True)
 data2.set_index("FellowshipID", inplace = True)
```

```python
data1.join(data2, lsuffix = "_left", rsuffix = "_right", how = "outer")
```

```python
data1.reset_index(inplace = True)
data2.reset_index(inplace = True)
```

```python
pd.concat([data1, data2])
```

## **Data Cleaning In Pandas**

```python
ccdata = pd.read_excel(r"C:\Users\sarvesh jathar\Downloads\Customer Call List.xlsx")
ccdata
```

```python
ccdata.info()
```

```python
ccdata = ccdata.drop_duplicates()
ccdata
```

```python
ccdata = ccdata.drop(columns = "Not_Useful_Column", axis = 1)
```

```python
ccdata
```

```python
ccdata["Last_Name"] = ccdata["Last_Name"].str.strip("123._/")
```

```python
ccdata
```

```python
ccdata["Phone_Number"] = ccdata["Phone_Number"].str.replace("[^a-zA-Z0-9]", "", regex=True)
```

```python
ccdata["Phone_Number"] = ccdata["Phone_Number"].apply(lambda x: str(x))
```

```python
ccdata["Phone_Number"] = ccdata["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:])
```

```python
ccdata
```

```python
ccdata["Phone_Number"] = ccdata["Phone_Number"].str.replace("Na", "")
ccdata["Phone_Number"] = ccdata["Phone_Number"].str.replace("nan--", "")
```

```python
ccdata
```

```python
ccdata[["Street", "State", "ZipCode"]] = ccdata["Address"].str.split(",", n = 2, expand = True)
```

```python
ccdata
```

```python
ccdata.drop(columns = "Address", inplace = True)
```

```python
ccdata
```

```python
ccdata["Do_Not_Contact"] = ccdata["Do_Not_Contact"].str.replace("Yes", "Y")
ccdata["Do_Not_Contact"] = ccdata["Do_Not_Contact"].str.replace("No", "N")
```

```python
ccdata["Paying Customer"] = ccdata["Paying Customer"].str.replace("Yes", "Y")
ccdata["Paying Customer"] = ccdata["Paying Customer"].str.replace("No", "N")
```

```python
ccdata
```

```python
ccdata = ccdata.replace("N/a", "")
ccdata = ccdata.fillna("")
```

```python
ccdata
```

```python
ccdatan = ccdata[ccdata["Do_Not_Contact"] == "N"]
ccdatan
```

```python
ccdatan[ccdatan["Phone_Number"] == ""]
```
